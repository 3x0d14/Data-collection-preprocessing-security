# Sentiment Analysis
Here we will use the library SentimentIntensityAnalyzer to perform the Sentiment Analysis, where for each tweet we will store the sentiment values in the following columns : 'negative' , 'positive' , 'neutrale' , 'compound' of the sentiment_data DataFrame. The DataFrame is concatenated with the original DataFrame of tweets.

```python
import preprocessor as pre
```

```python
pre.set_options(pre.OPT.URL,pre.OPT.MENTION,pre.OPT.EMOJI)
```

```python
for i in range(data.shape[0]):
    data["text"][i] = pre.clean(data["text"][i])
    if data["text"][i] == "":
        data.drop(i,inplace=True)
data
    
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
sentiment_results = []
analyser = SentimentIntensityAnalyzer()
for i in range(data.shape[0]):
    inter = {}
    inter["id"] = i
    sentiment = analyser.polarity_scores(data["text"][i])
    inter["negative"],inter["positive"],inter["neutrale"],inter["compound"] = sentiment.values()
    sentiment_results.append(inter)
```

```python
sentiment_data = pd.DataFrame(sentiment_results)
```

```python
sentiment_data.drop("id",axis=1,inplace=True)
sentiment_data
```

```python
data = pd.concat([data,sentiment_data],axis=1)
data
```
